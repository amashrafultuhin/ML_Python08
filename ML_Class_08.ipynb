### <font color = 'green'>Import Necessary Laibrary<font>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# for spliting test and train data
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#Mean Squared Error
from sklearn.metrics import mean_squared_error

#Mean Absolute Error
from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv("Height-Weight Data.csv")
```

```python
df.head(2)
```

```python
x = df[['Height']] #Independent variable
```

```python
y = df.Weight #Dependent variable
```

### <font color ='red'>Split test and train dataset</font>

```python
xtrain,xtest, ytrain,ytest = train_test_split(x,y, test_size = .3)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.shape[0]/xtrain.shape[0]
```

```python
reg = LinearRegression()
```

### <font color = 'green'>Train Linear regression model with train dataset</font>

```python
reg.fit(xtrain, ytrain)
```

```python
m = reg.coef_
m
```

```python
c = reg.intercept_
c
```

```python
y = (73.847017*m)+c
y
```

```python
reg.predict([[73.847017]])
```

```python
x.mean()
```

```python
y.mean()
```

```python
df
```

```python
df['Predicted Weight'] = reg.predict(x)
```

```python
df.head()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(df['Height'], df['Weight'], color='red')
plt.xlabel('Height', size=15)
plt.ylabel('Weight', size=15)
plt.xticks(size=15)
plt.yticks(size=15)

plt.show()
```

```python
plt.figure(figsize=(25,15))

plt.scatter(df['Height'],
            df['Weight'],
            color='red')

plt.plot(df['Height'],
         df['Predicted Weight'],
         'go--',
         linewidth=2,
         markersize=12)

plt.xlabel('Height',
           size=25)

plt.ylabel('Weight',
           size=25)

plt.xticks(size=25)

plt.yticks(size=25)

plt.grid()

plt.show()
```

```python
pred_y = reg.predict(xtest)
pred_y
```

```python
#Mean Sqyared Error
mse = mean_squared_error(ytest, pred_y)
```
